```python
html_doc = open("Dobiscuit.html", "r")
```

```python
from bs4 import BeautifulSoup
import re
soup = BeautifulSoup(html_doc)

print(soup.prettify())
```

```python
import os
```

```python
link_list = soup.find_all(href=re.compile("/pks/data_download/download/"))
page = "http://www.bio.nite.go.jp/"
filter_list = []
download_list = []

for link in link_list:
        l = link.get('href')
        filter_list.append(str(l))
print len(filter_list)
```

```python
download_list = []

for i in range(len(filter_list)):
    download_list.append(filter_list[i])
    ident1 = filter_list[i][28:68]
    m1 = re.search(ident1, filter_list[i])
    m2 = re.search(ident1, filter_list[i-1])    
    if type(m1) == type(m2):
        download_list.pop()
    else:
        pass
print len(download_list)
```

```python
for l in download_list:
    page = "http://www.bio.nite.go.jp"
    url = page+str(l)
    ident2 = url[-40:]
    #print url , ident2
    os.system("curl %s > %s.fasta" %(url, ident2))
    f = open("%s.fasta" % ident2, "r")
    r = f.readline()
    s = r.replace(" ","_")
    f.close()
    os.system("mv %s.fasta %s" % (ident2, s[1:]))
```
