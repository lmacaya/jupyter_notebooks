```python
%pylab inline
import prody as pd
import pylab
pylab.ion()
```

```python
structure = pd.parsePDB("/home/lmacaya/Escritorio/hGR/hGR_bio3d/hGR_eq.pdb")
trj = pd.Trajectory("/home/lmacaya/Escritorio/hGR/hGR_bio3d/hGR_800/hGR_popcwi_md_200ns_protein.dcd")
```

```python
trj.link(structure)
trj.setCoords(structure)
trj.setAtoms(structure.calpha)
```

```python
residue = structure['A', 276]
residue_phi, residue_psi, rgyr  = pylab.zeros(len(trj)), pylab.zeros(len(trj)), pylab.zeros(len(trj))

for i,frame in enumerate(trj):
    rgyr[i] = pd.calcGyradius(frame)
    residue_psi[i] = pd.calc(residue)
    residue_phi[i] = pd.calcPhi(residue)
```

```python
pylab.plot(residue_psi)
pylab.xlabel("Frame Index")
pylab.ylabel("Psi Angle")
```

```python
pylab.plot(residue_phi)
pylab.xlabel("Frame Index")
pylab.ylabel("Phi Angle")
```
