```python
def protein_hollow(model, hollow_output):
    
    model_memb = '%s%s_memb.pdb' % (hollow_output, model[:-4])
    model_hollow = '%s%s_hollow.pdb' % (hollow_output, model[:-4])
    
    #Generar archivo con membrana para modelos
    pymol.cmd.load(model_memb)
    pymol.cmd.load(model_hollow)
    pymol.cmd.remove('resname DUM')
    pymol.cmd.select('prot_hollow','%s_memb + %s_hollow' % (model[:-4], model[:-4]))
    pymol.cmd.save('%s_all.pdb' % model[:-4], 'prot_hollow', -1, 'pdb')
    return
```

```python
def lining_residues(model, hollow_output):
    
    model_memb = '%s%s_memb.pdb' % (hollow_output, model[:-4])
    model_hollow = '%s%s_hollow.pdb' % (hollow_output, model[:-4])
    
    #Generar archivo con residuos expuestos en canal central
    pymol.cmd.load(model_memb)
    pymol.cmd.load(model_hollow)
    pymol.cmd.remove('resname DUM')
    pymol.cmd.select('lining','byres %s_hollow around 3' % model[:-4])
    pymol.cmd.remove('solvent')
    pymol.cmd.save('%s_lining.pdb' % model[:-4], 'lining', -1, 'pdb')
    pymol.cmd.remove('all')
    
    #Filtro archivo pdb y obtenciÃ³n de residuos expuestos del canal
    
def parse_lining_residues(model, hollow_output):
    
    model_lining_residues_dict = {}
    
    import prody as pd
    
    model_lining_residues = '%s%s_lining.pdb' % (hollow_output, model[:-4])
    p = pd.parsePDB(model_lining_residues)
    alpha_carbons = p.select('calpha')
    
    lining_res = list(alpha_carbons.getResnames())
    lining_num_res = list(alpha_carbons.getResnums())
    lining_res_coord = list(alpha_carbons.getCoords())
    
    return lining_res, lining_num_res, lining_res_coord
```

```python
def residue_coords (residue, model_dict, lining_res_dict):

    residue_coords = []

    for item in model_dict.keys():
        for x,res in enumerate(lining_res_dict[item][0]):
            if res == '%s' % residue:
                residue_coords.append(list(lining_res_dict[item][2][x]))
       
    return residue_coords
```

```python
def plot_residue_coords (residue, residue_coords):
    import numpy as np
    from mpl_toolkits.mplot3d import Axes3D
    import matplotlib.pyplot as plt
    import matplotlib.mlab as mlab

    fig = plt.figure()

    ax = fig.add_subplot(111, projection='3d')
    
    x, y, z = np.array([i[0] for i in residue_coords]), np.array([i[1] for i in residue_coords]), np.array([i[2] for i in residue_coords])
    color = np.random.rand(len(residue_coords))
    ax.scatter(x, y, z, c=color, alpha=0.5)
    
    cset = ax.scatter(x, y, zdir='z', c='black', alpha=0.3)
    ax.set_title('%s' % residue)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_zlim(-50,300)
    
    fig.savefig('%s_coords.png' % residue, format='png', dpi=1000)
    fig.savefig('%s_coords.svg' % residue, format='svg', dpi=1000)

    plt.show()
    return
```
