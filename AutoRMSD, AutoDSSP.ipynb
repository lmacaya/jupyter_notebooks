# Cáculo RMSD con Pymol respecto a 4PYP

```python
#Importar Pymol
import __main__
__main__.pymol_argv = ['pymol','-qc'] # Pymol: quiet and no GUI
from time import sleep
import pymol
pymol.finish_launching()
```

```python
def do_rmsd(model, template, model_dir):
    try:
        template = template.lower()
        pymol.cmd.load('./%s.pdb' % template) #Cargar molde
    except:
        pymol.cmd.fetch('%s' % template)
    pymol.cmd.load('%s%s' % (model_dir, model)) #Cargar modelo
    RMSD = pymol.cmd.cealign(template, '%s' % model[:-4]) #Cálculo RMSD
    return RMSD['RMSD'], RMSD['alignment_length'] #Obtener valores de RMSD y residuos alineados
```

```python
def do_dssp(model, template, model_dir):
    import os
    
    dssp_bin = "/usr/local/bin/mkdssp"
    command = "%s %s%s %s.dssp" % (dssp_bin, model_dir, model, model[:-4]); #Ejecucción DSSP
    os.system(command)
    
    dssp_file = open("%s.dssp"% model[:-4], 'r')
    lineas = dssp_file.readlines()
    
    dssp_info = ""
    counter = 0

    #Obtener columna con denominación para estructura secundaria
    for num in range(28,len(lineas)):
        if lineas[num][16] == " ":
            dssp_info += "-"
        if lineas[num][16] != " ":
            dssp_info += lineas[num][16]
            counter += 1

    return dssp_info 
```

```python
def percentage_ss (dssp, template, model_dir):
    
    #Contar residuos según denominación de estructura secundaria
    counter_helix = 0
    counter_beta = 0
    counter_turns = 0
    counter_bends = 0
    counter_loops = 0
    
    for letter in dssp:
        if letter == "G" or letter == "H" or letter == "I":
            counter_helix += 1
        elif letter == "B" or letter == "E":
            counter_beta += 1
        elif letter == "T":
            counter_turns += 1
        elif letter == "S":
            counter_bends += 1
        else:
            counter_loops += 1
    
    return counter_helix, counter_beta, counter_turns, counter_bends, counter_loops
```
